```python
def removeDupesAndNulls(df):
    df = df.dropna()
    df = df.drop_duplicates()
    return df
```

```python
def removeOutliers(df):
    mask = pd.Series(True, index=df.index)
    numericalColumns = df.select_dtypes(include='number')
    for column in numericalColumns:
        Q1 = df[f'{column}'].quantile(0.25)
        Q3 = df[f'{column}'].quantile(0.75)
        IQR = Q3 - Q1 

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        mask &= df[column].between(lower, upper)

    clean_df = df[mask]
    outliers_df = df[~mask]

    return clean_df, outliers_df
```

```python
df_titanic = sns.load_dataset("titanic")
```

```python
removeDupesAndNulls(df_titanic)
```

```python
clean_df, outliers_df = removeOutliers(df_titanic)
```
